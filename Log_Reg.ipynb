- Learning rate in ML: We have to make transformation on our line to correctly
classify our outcomes, if we do that witout learning rate then our transformation would show some drastic changes on our line so we use small learning rate which minimize our transformation to prevent overshoot of our model.

```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=10000, n_features=2, n_informative=1, n_redundant=0, random_state=41,n_clusters_per_class=1)
print(X.shape, y.shape)
```

```python
 import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=25, edgecolor='k')
plt.show()
```

```python
def perceptron(X,y):
    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])
    for i in range(10000):
        j=np.random.randint(0,len(X))
        y_hat=step(np.dot(weights,X[j]))
        weights=weights+0.1*(y[j]-y_hat)*X[j]
    return weights[0],weights[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_=perceptron(X,y)
```

```python
m=-coef_[0]/coef_[1]
c=-intercept_/coef_[1]
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=25, edgecolor='k')
plt.plot(X[:, 0], m*X[:, 0] + c, color='red')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(X,y)
m1=logreg.coef_[0][0]
c1=logreg.intercept_[0]
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=25, edgecolor='k')
plt.plot(X[:, 0], m1*X[:, 0] + c1, color='red')
plt.plot(X[:, 0], m*X[:, 0] + c, color='green')
plt.show()
```
